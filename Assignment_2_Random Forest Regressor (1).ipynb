METHOD 1: SUPPORT VECTOR MACHINE FOR REGRESSION ANALYSIS

Since the task identified is a regression task based on the nature of target variable, we will perform Support Vector Regression(SVR). It is of two types:

- Linear SVR
- Kernel SVR where kernel can be linear,polynomial, or Gaussian.

Steps followed to do Support Vector Regression are:

1. Necessary packages are imported
2. Daatset is splitted into training and test data in the ratio of 80:20.
3. Both training and test data is standardised using StandardScaler().
4. Build and evaluate Linear SVR model.
5. Build and evaluate SVR model with RBF kernel.
6. Build and evaluate SVR with polynomial kernel.
7. Build and evaluate SVR with linear kernel.
8. Summarise and select the best model of SVM.

```python
#Importing necessary packages
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.svm import LinearSVR
from sklearn.metrics import mean_squared_error

```

```python
#Splitting data into training and test data
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
#Standardising train and test data
sx=StandardScaler()
sy=StandardScaler()
Xtrain_scaled=sx.fit_transform(Xtrain)
Xtest_scaled=sx.transform(Xtest)
ytrain_scaled=sy.fit_transform(ytrain)
ytest_scaled=sy.transform(ytest)
```

1.1 Linear Support Vector Regression(SVR)

Here,

Step1: We take a range of epsilon values.

Step2: Create an empty list of test errors.

Step3: For each epsilon value in the range defined,
  - we fit the LinearSVR() model
  - Make prediction on unseen scaled data (test scaled data)
  - Transform the predicted values back to the original scale
  - Calculate mean squared error
  - Append each calculated mean squared error to the empty list
Step4: Find minimum error and its associated epsilon value

step5: Plot the results

```python
epsilon=[0.1,0.5,1.0,1.5] #range of epsilon values
test_error_linear=[] #empty list of errors
for e in epsilon:
    #Fitting a model
    svm_linear=LinearSVR(epsilon=e,dual=True,random_state=42)
    svm_linear.fit(Xtrain_scaled,ytrain_scaled)
    #Predicting y values
    ypred_scaled=svm_linear.predict(Xtest_scaled)
    ypred_scaled=ypred_scaled.reshape(-1,1)
    ypred=sy.inverse_transform(ypred_scaled)
    #Calculating prediction error
    mse=mean_squared_error(ytest,ypred)
    print(f'Epsilon= {e},Test Error= {mse}')
    #Appending each calculated error value to the empty list
    test_error_linear.append(mse)

```

```python
min_error_index=np.argmin(test_error_linear)
print("Epsilon value with minimum test error is:",epsilon[min_error_index])
print("Minimum Test Error is",test_error_linear[min_error_index])
```

```python
# Plot the results
plt.plot(epsilon,test_error_linear, marker='o')
plt.title('Linear SVR with Different Epsilon Values')
plt.xlabel('Epsilon')
plt.ylabel('Mean Squared Error')
plt.axis([0,2,0,17])
plt.show()
```

So a Linear SVR with epsilon value of 0.1 perfforms well. the plot clearly indicates that with the increasing epsilon values the test error also increases.

1.2 SVR with RBF Kernel

```python
#fitting the model with rbf kernel
C=[0.01,0.1,1,10,100]  #Setting a range of regularization parameter
rbf_error=[] #empty list of errors
for c in C:
    #fitting model for different epsilon values
    regressor = SVR(kernel = 'rbf',C=c)
    regressor.fit(Xtrain_scaled, ytrain_scaled)
    #Making Predictions
    y_pred = regressor.predict(Xtest_scaled)
    y_pred =y_pred.reshape(-1,1)
    ypred = sy.inverse_transform(y_pred)
    #Checking for test error
    mse=mean_squared_error(ytest,ypred)
    print(f'Regularization Parameter= {c},RBF Test Error= {mse}')
    rbf_error.append(mse)
```

```python
# Plot the results
plt.plot(C,rbf_error, marker='o')
plt.title('SVR with RBF kernel and Different Hyperparameter Values')
plt.xlabel('Regularisation Strength')
plt.ylabel('RBF Test error')
plt.axis([-1,100,0,40])
plt.show()
```

It can be seen that for larger hyperparameter values, the test error decreases and improves the model accuracy in making prediction.

```python
ind=np.argmin(rbf_error)
print(f'Minimum Test error is {rbf_error[ind]} at C={C[ind]}')
```

1.3 SVR with Polynomial Kernel

```python
degrees=10
poly_error=[]
for d in range(degrees):
    svr_poly=SVR(kernel='poly',degree=d)
    svr_poly.fit(Xtrain_scaled,ytrain_scaled)
    poly_ypred=svr_poly.predict(Xtest_scaled)
    poly_ypred=poly_ypred.reshape(-1,1)
    Ypred=sy.inverse_transform(poly_ypred)
    poly_mse=mean_squared_error(ytest,Ypred)
    print(f'Degree={d},Polynomial Model error={poly_mse}')
    poly_error.append(poly_mse)
```

A model with polynomial degree of 1 has the least test error. So, next we will again build a model with polynomial kernel of degree 1 and range oc hyperparameters to see if the model can be improved.

```python
C=[0.01,0.1,1,10,100]
poly_error=[]
for c in C:
    svr_poly=SVR(kernel='poly',C=c,degree=1)
    svr_poly.fit(Xtrain_scaled,ytrain_scaled)
    poly_ypred=svr_poly.predict(Xtest_scaled)
    poly_ypred=poly_ypred.reshape(-1,1)
    Ypred=sy.inverse_transform(poly_ypred)
    poly_mse=mean_squared_error(ytest,Ypred)
    print(f'Regularisation Strength={c},Test error={poly_mse}')
    poly_error.append(poly_mse)
```

```python
# Plot the results
plt.plot(C, poly_error, marker='*')
plt.title('SVR with Polynomial Kernel and Different C Values')
plt.xlabel('C Value')
plt.ylabel('Mean Squared Error')
plt.xscale('log')  # Use a logarithmic scale for better visualization
plt.show()
```

```python
pos=np.argmin(poly_error)
print(f'Minimum Test Error for polynomial degree 1 is: {poly_error[pos]} at C={C[pos]}')
```

1.3 Support vector Regression with Linear Kernel

```python
C=[0.01,0.1,1,10,100]  #Setting a range of regularization parameter
test_error=[] #empty list of test errors
for c in C:
    svr=SVR(kernel="linear",C=c)
    svr.fit(Xtrain_scaled,ytrain_scaled)
    predicted_y=svr.predict(Xtest_scaled)
    predicted_y=predicted_y.reshape(-1,1)
    pred_y=sy.inverse_transform(predicted_y)
    te=mean_squared_error(ytest,pred_y)
    print(print(f'C={c}, Test Error={te}'))
    test_error.append(te)

```

```python
index=np.argmin(test_error)
print(f'Minimum test error is {test_error[index]} for hyperparameter value at {C[index]}')
```

```python
# Plot the results
plt.plot(C, test_error, marker='*')
plt.title('SVR with Linear Kernel and Different C Values')
plt.xlabel('C Value')
plt.ylabel('Mean Squared Error')
plt.xscale('log')  # Use a logarithmic scale for better visualization
plt.show()
```

Except for hyperparameter value of 0.01, we are getting the same test error vale for all other values.

MSE measures how well the model's predictions match the actual target values. A lower MSE indicates better model performance.

R-squared is a measure of how well the model explains the variance in the data. It ranges from 0 to 1, where a higher R-squared indicates a better fit.
